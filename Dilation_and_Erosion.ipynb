```python
# Import OpenCV for image processing
import cv2
# Import Numpy for Numerical operations
import numpy as np
# Import matplotlib for displaying images
import matplotlib.pyplot as plt

# Load image
image = cv2.imread("bird.jpg")

# Check if image loaded correctly
if image is None:
    print("Error: Image not found.")
else:
    print("Image loaded successfully.")

# Convert from BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display original image
plt.figure(figsize=(6,6))
plt.imshow(image_rgb)
plt.title("Orginal Image")
plt.axis("off")
plt.show()
```

```python
# Convert image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Save image
cv2.imwrite("grayscale.jpg", gray_image)

# Display grayscale image
plt.figure(figsize=(6,6))
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
# Apply thresholding
# Pixels greater than 127 become 255 (white)
# Pixels less than 127 become 0 (black)
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

# Save image
cv2.imwrite("binary.jpg", binary_image)

# Display binary image
plt.figure(figsize=(6,6))
plt.imshow(binary_image, cmap='gray')
plt.title("Binary Image")
plt.axis("off")
plt.show()
```

```python
# Create a 5x5 structuring element (kernel)
# This is a square kernel filled with 1s
kernel = np.ones((5,5), np.uint8)

# Print kernel to see its structure
print("Structuring Element (Kernel):")
print(kernel)
```

```python
# Apply erosion
# The kernel slides over the image
# White pixels remain only if kernel fits entirely
eroded_image = cv2.erode(binary_image, kernel, iterations=1)

# Save image
cv2.imwrite("eroded.jpg", eroded_image)

# Display binary image
plt.figure(figsize=(6,6))
plt.imshow(eroded_image, cmap='gray')
plt.title("Eroded Image")
plt.axis("off")
plt.show()
```

```python
# Apply dilation
# White pixels expand
dilated_image = cv2.dilate(binary_image, kernel, iterations=1)

# Save image
cv2.imwrite("dilated.jpg", dilated_image)

# Display binary image
plt.figure(figsize=(6,6))
plt.imshow(dilated_image, cmap='gray')
plt.title("Dilated Image")
plt.axis("off")
plt.show()
```

```python
# Show all images together for comparison
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(image_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(binary_image, cmap='gray')
plt.title("Binary")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(eroded_image, cmap='gray')
plt.title("Erosion")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(dilated_image, cmap='gray')
plt.title("Dilation")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Opening -> Erosion to Dilation
opening = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, kernel)

# Save image
cv2.imwrite("opening.jpg", opening)

plt.imshow(opening, cmap='gray')
plt.title("Opening (Erosion + Dilation)")
plt.axis("off")
plt.show()
```

```python
# Closing -> Dilation to Erosion
closing = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel)

# Save image
cv2.imwrite("closing.jpg", closing)

plt.imshow(closing, cmap='gray')
plt.title("Closing (Dilation + Erosion)")
plt.axis("off")
plt.show()
```
